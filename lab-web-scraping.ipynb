```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin


BASE_URL = "https://books.toscrape.com/"
CATALOGUE_URL = urljoin(BASE_URL, "catalogue/page-1.html")

RATING_MAP = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5,
}

COLUMNS = [
    "UPC",
    "Title",
    "Price (£)",
    "Rating",
    "Genre",
    "Availability",
    "Description",
]


def get_soup(url, session):
    """Download a page and parse it with BeautifulSoup."""
    response = session.get(url, timeout=15)
    response.raise_for_status()
    response.encoding = "utf-8"
    return BeautifulSoup(response.text, "html.parser")


def parse_price(price_text):
    numeric_price = "".join(
        character for character in price_text if character.isdigit() or character == "."
    )
    return float(numeric_price)


def parse_rating(parent):
    rating_tag = parent.select_one("p.star-rating")
    rating_word = next(
        class_name for class_name in rating_tag["class"] if class_name != "star-rating"
    )
    return RATING_MAP[rating_word]


def parse_product_table(detail_soup):
    product_info = {}

    for row in detail_soup.select("table.table tr"):
        header = row.find("th")
        value = row.find("td")

        if header and value:
            product_info[header.get_text(strip=True)] = value.get_text(" ", strip=True)

    return product_info


def parse_genre(detail_soup):
    breadcrumb_links = [
        link.get_text(strip=True) for link in detail_soup.select("ul.breadcrumb li a")
    ]
    return breadcrumb_links[-1] if len(breadcrumb_links) >= 3 else "Unknown"


def parse_description(detail_soup):
    description_header = detail_soup.select_one("#product_description")

    if description_header:
        description = description_header.find_next_sibling("p")
        if description:
            return description.get_text(" ", strip=True)

    return ""


def parse_book_details(detail_url, session):
    detail_soup = get_soup(detail_url, session)
    product_info = parse_product_table(detail_soup)

    return {
        "UPC": product_info.get("UPC", ""),
        "Title": detail_soup.select_one("div.product_main h1").get_text(strip=True),
        "Genre": parse_genre(detail_soup),
        "Availability": product_info.get("Availability", ""),
        "Description": parse_description(detail_soup),
    }


def scrape_books(min_rating, max_price):
    """
    Scrape Books to Scrape and return books with rating >= min_rating
    and price <= max_price.
    """
    if not 1 <= min_rating <= 5:
        raise ValueError("min_rating must be between 1 and 5.")
    if max_price < 0:
        raise ValueError("max_price must be 0 or greater.")

    session = requests.Session()
    session.headers.update({"User-Agent": "Ironhack web scraping lab"})

    books = []
    page_url = CATALOGUE_URL

    while page_url:
        listing_soup = get_soup(page_url, session)

        for book in listing_soup.select("article.product_pod"):
            price = parse_price(book.select_one("p.price_color").get_text(strip=True))
            rating = parse_rating(book)

            if rating >= min_rating and price <= max_price:
                detail_path = book.select_one("h3 a")["href"]
                detail_url = urljoin(page_url, detail_path)
                details = parse_book_details(detail_url, session)

                books.append(
                    {
                        "UPC": details["UPC"],
                        "Title": details["Title"],
                        "Price (£)": price,
                        "Rating": rating,
                        "Genre": details["Genre"],
                        "Availability": details["Availability"],
                        "Description": details["Description"],
                    }
                )

        next_page = listing_soup.select_one("li.next a")
        page_url = urljoin(page_url, next_page["href"]) if next_page else None

    return pd.DataFrame(books, columns=COLUMNS)


books_df = scrape_books(min_rating=4.0, max_price=20.0)
print(f"Found {len(books_df)} books with rating >= 4.0 and price <= £20.00")
books_df
```
